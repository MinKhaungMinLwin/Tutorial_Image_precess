## creating img with numpy

```python
import numpy as np

def main():
    img = np.zeros((512,512,3),np.uint8)
    
    cv.line(img,(0,0),(100,100),(255,0,0),2)
    
    cv.imshow('Black',img)
    cv.waitKey(0)
    cv.destroyAllWindows()
    
if __name__ == "__main__":
    main()
```

```python
def main():
    img = np.zeros((512,512,3),np.uint8)
    
    cv.rectangle(img,(40,60),(80,100),(0,255,0),2)
    
    cv.circle(img,(150,150),20,(0,0,255),2)
    
    cv.ellipse(img,(200,200),(50,20),0,0,360,(255,0,0),2)
    
    points = np.array([[67,88],[80,90],[99,55]])
    print(points)
    
    cv.polylines(img,[points],False,(0,255,0))
    
    text = "ROM Robotics"
    
    cv.putText(img,text,(100,400),cv.FONT_HERSHEY_SIMPLEX,1.2,(255,255,0))
    
    cv.imshow('Black', img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

## creating a trackbar

```python
def Bfunc(value):
    pass

def Gfunc(value):
    pass

def Rfunc(value):
    pass

def main():
    img = np.zeros((512,512,3), np.uint8)
    
    window = "OpenCV Color Plattee"
    cv.namedWindow(window)
    
    cv.createTrackbar('B',window,0,255,Bfunc)
    cv.createTrackbar('G',window,0,255,Gfunc)
    cv.createTrackbar('R',window,0,255,Rfunc)
    
    while True:
        cv.imshow(window,img)
        
        blue = cv.getTrackbarPos('B',window)
        green = cv.getTrackbarPos('G',window)
        red = cv.getTrackbarPos('R',window)
        
        img[:] = [blue,green,red]
        print(blue,green,red)
        
        if cv.waitKey(1) == 27:
            break
        
    cv.destroyAllWindows()
    
if __name__ == "__main__":
    main()
```

```python
drawing = False
mode = True

(ix,iy) = (-1,-1)

window_name = "Drawing"
cv.namedWindow(window_name)

img = np.zeros((512,512,3),np.uint8)

def draw(event,x,y, flags, param):
    global drawing,mode,ix,iy
    
    if event == cv.EVENT_LBUTTONDOWN:
        drawing = True
        (ix,iy) = x,y
    elif event == cv.EVENT_MOUSEMOVE:
        if drawing == True:
            if mode == True:
                cv.rectangle(img,(ix,iy),(x,y),(255,0,0),-1)
            else:
                cv.circle(img,(x,y),30,(0,255,0),-1)
                
    elif event == cv.EVENT_LBUTTONUP:
        drawing = False
        
cv.setMouseCallback(window_name,draw)

def main():
    global mode
    while(True):
        cv.imshow(window_name,img)
        k = cv.waitKey(1)
        if k == ord('m') or k == ord('M'):
            mode = not mode
        elif k == 27:
            break
    cv.destroyAllWindows()
    
if __name__ == "__main__":
    main()
```

## read webcam

```python
def main():
    cap = cv.VideoCapture(0)
    
    ret, frame = cap.read()
    
    if cap.isOpened():
        ret, frame = cap.read()
        print(ret)
        
    img = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
    
    plt.imshow(img)
    plt.show()
    
    cap.release()

if __name__ == "__main__":
    main()
```
